```python
#modelagem
import pandas as pd
import json
from flatten_json import flatten

#requisições para API
import requests as r

#visualização de dados
import plotly.express as px
```

```python
#mostrar todas as colunas do df
pd.set_option('display.max_columns', None)
```

```python
#acessando e consumindo dados da url através de API do IBGE
url = "https://servicodados.ibge.gov.br/api/v3/agregados/6579/periodos/2001|2002|2003|2004|2005|2006|2008|2009|2011|2012|2013|2014|2015|2016|2017|2018|2019|2020|2021/variaveis/9324?localidades=N1[all]"
headers = {
    'Content-Type': 'application/json;charset=UTF-8',
    'User-Agent': 'google-colab',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
}
requisicao = r.get(url, headers=headers)
```

```python
#verificando a situação dos serviços solicitado pela API
if requisicao.status_code == 200:
  print('Requisição atendida')
  dados_requisicao = requisicao.json()
else:
    print('Requisição Off')
```

```python
#verificando a estrutura do json com o comando for
for i in dados_requisicao:
  print(f'Inforação ->{dados_requisicao}')
```

```python
#se não fizer o nivelamento do arquivo json 
#com a biblioteca flatten_json veja o oque 
#ocorre ao transformar os dados em um DF
df_sem = pd.DataFrame(dados_requisicao)
df_sem.head(2)
```



---



```python
#nivelando o formato do json com biblioteca flatten_json
#para cada chave ela cria uma nova coluna
dados_dic_flattened = [flatten(d) for d in dados_requisicao]
```

```python
#Atribuindo os dados nivelados ao dataFrame
df_dados = pd.DataFrame(dados_dic_flattened)
```

```python
df_dados.head()
```

```python
#retirando as colunas que não preciso
df_dados.drop(["id"], axis=1, inplace=True)
df_dados.drop(["variavel"], axis=1, inplace=True)
df_dados.drop(["unidade"], axis=1, inplace=True)
df_dados.drop(["resultados_0_classificacoes"], axis=1, inplace=True)
df_dados.drop(["resultados_0_series_0_localidade_id"], axis=1, inplace=True)
df_dados.drop(["resultados_0_series_0_localidade_nivel_id"], axis=1, inplace=True)
df_dados.drop(["resultados_0_series_0_localidade_nivel_nome"], axis=1, inplace=True)
#df_dados.drop(["resultados_0_series_0_serie_2001"], axis=1, inplace=True)
#df_dados.drop(["resultados_0_series_0_serie_2002"], axis=1, inplace=True)
#df_dados.drop(["resultados_0_series_0_serie_2003"], axis=1, inplace=True)
#df_dados.drop(["resultados_0_series_0_serie_2004"], axis=1, inplace=True)
```

```python
#renomeando as colunas
df_dados = df_dados.rename(columns={'resultados_0_series_0_localidade_nome': 'Localidade',
                                    'resultados_0_series_0_serie_2001': 'Ano 2001',
                                    'resultados_0_series_0_serie_2002': 'Ano 2002',
                                    'resultados_0_series_0_serie_2003': 'Ano 2003',
                                    'resultados_0_series_0_serie_2004': 'Ano 2004',
                                    'resultados_0_series_0_serie_2005': 'Ano 2005', 
                                    'resultados_0_series_0_serie_2006': 'Ano 2006',
                                    'resultados_0_series_0_serie_2007': 'Ano 2007',
                                    'resultados_0_series_0_serie_2008': 'Ano 2008',
                                    'resultados_0_series_0_serie_2009': 'Ano 2009',
                                    'resultados_0_series_0_serie_2010': 'Ano 2010',
                                    'resultados_0_series_0_serie_2011': 'Ano 2011',
                                    'resultados_0_series_0_serie_2012': 'Ano 2012',
                                    'resultados_0_series_0_serie_2013': 'Ano 2013',
                                    'resultados_0_series_0_serie_2014': 'Ano 2014',
                                    'resultados_0_series_0_serie_2015': 'Ano 2015',
                                    'resultados_0_series_0_serie_2016': 'Ano 2016',
                                    'resultados_0_series_0_serie_2017': 'Ano 2017',
                                    'resultados_0_series_0_serie_2018': 'Ano 2018',
                                    'resultados_0_series_0_serie_2019': 'Ano 2019',
                                    'resultados_0_series_0_serie_2020': 'Ano 2020',
                                    'resultados_0_series_0_serie_2021': 'Ano 2021',
                                    'resultados_0_series_0_localidade_nome': 'Pais'                                    })
```

```python
df_dados
```

```python
df_br = pd.DataFrame(df_dados.transpose()).reset_index()
```

```python
df_br
```

```python
#apando a primeira linha
df_br.drop(df_br.index[0], inplace=True)
```

```python
#Renomeando as colunas
df_br.rename(columns={'index': 'Periodo', 0: 'Pop Estimada'}, inplace=True)
```

```python
#verificando o tipo dos dados
df_br.dtypes
```

```python
#convertendo a coluna
df_br['Pop Estimada'] = df_br['Pop Estimada'].astype('float64')
```

```python
#verificando o tipo dos dados se foi alterado
df_br.dtypes
```

# Aplicando Visualização dos dados e analise descritivas

```python
#Analisando a estatistica descritiva do df_br
df_br.describe()
```

```python
#criando uma nova coluna para mostrar os dados de comparação ao ano anterior
df_br['Populacao Anterior'] = df_br['Pop Estimada'].shift(1)
df_br.head()
```

```python
#calculando o percentual referente ao ano anterior
df_br['Dif_perc'] = (((df_br['Pop Estimada'] - df_br['Populacao Anterior']) / df_br['Populacao Anterior'])  * 1000)/100
df_br.head(3)
```

```python
#aplicando o boxsplot para identificar possiveis Outliers
import plotly.io as pio

pio.templates.default='plotly_dark'
fig = px.box(df_br, y=['Pop Estimada'])
fig.show()
```

```python
figsize=(11, 8)
fig = px.line(df_br, x=df_br["Periodo"], y=df_br["Pop Estimada"])

#personalizando o granfico
fig.update_layout(
    title= 'População Residente Estimada',    
    xaxis_title= 'Periodo',
    yaxis_title= 'População (em milhões)',    
    margin=dict(l=50, r=50, t=50, b=50),
    template="plotly_dark",
    #rodapé
    annotations= [dict(text='Fonte: IBGE', xref='paper', yref='paper', x=0, y=-0.10, showarrow=False)])

fig.show()
```

```python
#visualizando os dados no formato em percentual
figsize=(11, 8)
fig = px.line(
    df_br, x=df_br["Periodo"], 
    y=df_br["Dif_perc"]
    )

#personalizando o granfico
fig.update_layout(
    title= 'População Residente Estimada (Acompanhamento em %)',    
    xaxis_title= 'Periodo',
    yaxis_title= 'População (em milhões)',
    
    margin=dict(l=50, r=50, t=50, b=50),
    template="plotly_dark",
    #rodapé
    annotations= [dict(text='Fonte: IBGE', xref='paper', yref='paper', x=0, y=-0.10, showarrow=False)])

fig.show()
```
